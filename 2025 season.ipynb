```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.formula1.com/en/results/2025/races"
data = requests.get(url)
soup = BeautifulSoup(data.content, 'html.parser')
races_table = soup.select('table.f1-table')[0]
links = races_table.find_all('a', href=True)
links = [l.get("href") for l in links]
links = [l for l in links if '/2025/' in l]
```

```python
links
```

```python
race_urls = [f"https://formula1.com{l}" for l in links]
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
races = pd.read_html(str(races_table))[0]
races
```

```python
races = races.iloc[: -2]
```

```python
gp_rename = {
    'Flag of AustraliaAustralia': 'Australian Grand Prix',
    'Flag of People’s Republic of ChinaChina': 'Chinese Grand Prix',
    'Flag of MonacoMonaco': 'Monaco Grand Prix',
    'Flag of CanadaCanada': 'Canadian Grand Prix',
    'Flag of AustriaAustria': 'Austrian Grand Prix',
    'Flag of JapanJapan': 'Japanese Grand Prix',
    'Flag of BahrainBahrain': 'Bahrain Grand Prix',
    'Flag of Saudi ArabiaSaudi Arabia': 'Saudi Arabian Grand Prix',
    'Flag of United States of AmericaMiami': 'Miami Grand Prix',
    'Flag of ItalyEmilia-Romagna': 'Emilia Romagna Grand Prix',
    'Flag of SpainSpain': 'Spanish Grand Prix',
}

races['GRAND PRIX'] = races['GRAND PRIX'].replace(gp_rename)
races
```

```python
races['DATE'] = races['DATE'].apply(lambda x: f"{x} 2025")
races['DATE'] = pd.to_datetime(races['DATE'], format='%d %b %Y')
races.drop(columns=['WINNER', 'TEAM', 'LAPS', 'TIME'], inplace=True)
races
```

```python
race_result_urls = race_urls
```

```python
race_quali_urls = [url.replace('race-result', 'qualifying') for url in race_urls]
race_quali_urls = race_quali_urls[: -2]
race_result_urls = race_result_urls[: -2]
```

```python
print(race_quali_urls[-1]), print(race_result_urls[-1])
```

```python
results = []
for url in race_result_urls:
    data = requests.get(url)
    soup = BeautifulSoup(data.content, 'html.parser')
    results_table = soup.select('table.f1-table')[0]
    res = pd.read_html(str(results_table))[0]
    results.append(res)
results = pd.concat(results, ignore_index=True)
```

```python
results.reset_index
results.drop(['PTS.', 'NO.', 'LAPS'])
res_rename = {
    'Lando NorrisNOR': 'norris',
    'Max VerstappenVER': 'max_verstappen',
    'George RussellRUS': 'russel',
    'Flag of CanadaCanada': 'Canadian Grand Prix',
    'Flag of AustriaAustria': 'Austrian Grand Prix',
    'Flag of JapanJapan': 'Japanese Grand Prix',
    'Flag of BahrainBahrain': 'Bahrain Grand Prix',
    'Flag of Saudi ArabiaSaudi Arabia': 'Saudi Arabian Grand Prix',
    'Flag of United States of AmericaMiami': 'Miami Grand Prix',
    'Flag of ItalyEmilia-Romagna': 'Emilia Romagna Grand Prix',
    'Flag of SpainSpain': 'Spanish Grand Prix',
}
results['DRIVER'] = results['DRIVER'].replace(res_rename)
results
```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_Formula_One_circuits'
webpage = requests.get(url)
soup = BeautifulSoup(webpage.content, 'lxml')

table = soup.find('table', {'class': 'wikitable sortable'})
```

```python
headers = [th.get_text(strip=True) for th in table.find_all('th')]

rows = []
for tr in table.find_all('tr')[1:]:
    cells = tr.find_all('td')
    if len(cells) == len(headers):
        row = [cell.get_text(strip=True) for cell in cells]
        rows.append(row)

circuits_df = pd.DataFrame(rows, columns=headers)
circuits_df
```

```python
circuits_df['Circuit'] = circuits_df['Circuit'].str.replace('*', '', regex=False).str.strip()
circuits_df['Circuit'] = circuits_df['Circuit'].replace({
    'Albert Park Circuit': 'Albert Park Grand Prix Circuit',
    'Suzuka International Racing Course': 'Suzuka Circuit',
    'Circuit de Spa-Francorchamps': 'Circuit de Spa-Francorchamps',
    'Autodromo Internazionale Enzo e Dino Ferrari': 'Autodromo Enzo e Dino Ferrari',
    'Autodromo José Carlos Pace': 'Autódromo José Carlos Pace',
    'Autodromo Nazionale di Monza': 'Autodromo Nazionale di Monza',
    'Circuit Gilles-Villeneuve': 'Circuit Gilles Villeneuve',
    'Circuit Zandvoort': 'Circuit Park Zandvoort',
    'Las Vegas Strip Circuit': 'Las Vegas Strip Street Circuit',
    'Lusail International Circuit': 'Losail International Circuit',
    'Suzuka International Racing Course': 'Suzuka Circuit',
})
```

```python
circuits_df.rename(columns={
    'Circuit': 'circuit name'}, inplace=True)
```
